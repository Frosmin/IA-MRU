```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python

np.random.seed(42)
tiempos = np.random.uniform(0, 10, 1000)  
xo = 0
x = np.random.uniform(0,10,1000)


velocidades = x/tiempos





train_tiempos = tiempos
train_posiciones = x
train_velocidades = velocidades


# test_tiempos = tiempos[800:]
# test_posiciones = posiciones[800:]
```

```python
train_entradas = np.column_stack((train_posiciones, train_tiempos))

```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(2,)),  
    tf.keras.layers.Dense(1) 
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(train_entradas, train_velocidades, epochs=100)
```

```python


# # Mostrar algunas predicciones
# plt.scatter(test_tiempos, test_posiciones, color='blue', label='Datos reales')
# plt.scatter(test_tiempos, predicciones, color='red', label='Predicciones')
# plt.xlabel('Tiempo')
# plt.ylabel('Posición')
# plt.legend()
# plt.show()



def predecir_velocidad(distancia, tiempo):
    entrada = np.array([[distancia, tiempo]])
    prediccion = model.predict(entrada)
    return prediccion[0][0]

distancia = 10  
tiempo = 10     
velocidad_predicha = predecir_velocidad(distancia, tiempo)
print(f'Predicción de velocidad para distancia = {distancia} y tiempo = {tiempo}: {velocidad_predicha} m/s')
```
